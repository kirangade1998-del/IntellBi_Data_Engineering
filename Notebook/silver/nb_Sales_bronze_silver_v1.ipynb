```python
%sql


INSERT INTO  intellbi_catalog.intellbi_silver.cleansedMonthlySales
SELECT 
DISTINCT
 cast(trim(sale_id )as integer)AS sale_id,

 upper(trim(product)) AS product,
 upper(trim(category)) AS category,
cast(trim(quantity) as integer)as quantity,
cast(trim(price) as double)as price,
 to_date(trim(sale_date),'yyy-MM-dd')as sale_date,
 upper(trim(region)) AS region,
current_timestamp() as ingest_ts
FROM intellbi_catalog.intellbi_bronze.rawMonthlysales
WHERE ingest_ts>(SELECT coalesce(max(ingest_ts), TO_DATE ('1990-01-01','yyyy-MM-dd'))
from intellbi_catalog.intellbi_silver.cleansedMonthlySales)
 ORDER BY cast (trim(sale_id )as integer);
    




```
