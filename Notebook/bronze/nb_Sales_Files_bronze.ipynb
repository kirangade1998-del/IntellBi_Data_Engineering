```python
from pyspark.sql.functions import current_timestamp
dfMonthlySales = dfMonthlySales.selectExpr(
    "UPPER(TRIM(sale_id)) as sale_id",
    "UPPER(TRIM(product)) as product",
    "UPPER(TRIM(category)) as category",
    "TRIM(quantity) as quantity",
    "TRIM(price) as price",
    "TRIM(sale_date) as sale_date",
    "UPPER(TRIM(region)) as region"
    ).withColumn("ingest_ts", current_timestamp())
    


dfMonthlySales.display()

# dfMonthlySales.write.mode("overwrite").saveAsTable("intellbi_catalog.intellbi_bronze.dfMonthlySales")

```

```python
dfMonthlySales.printSchema()
```

```python
dfMonthlySales.write \
    .mode("overwrite") \
    .option("overwriteSchema", "true") \
    .saveAsTable("intellbi_catalog.intellbi_bronze.dfMonthlySales")
```
